# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randn(1000)

plt.hist (data, bins = 30, color = "pink", edgecolor = "purple", alpha = 0.7) # Histogram with 30 bins
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
