# Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 100 * np.random.rand(50)

plt.scatter(x, y, c = colors, s = sizes, alpha = 0.7, cmap = 'viridis') # Scatter plot with color map
plt.title("Scatter Plot")
plt.xlabel("Random x")
plt.ylabel("Random y")
plt.colorbar( )  # Show color scale
plt.grid(True)
plt.show( )
```
