# Bar Plot
## Vertical

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
show_categtories = ["LOTR", "Harry Potter", "Star Wars", "GOT", "LOST", "Friends", "Breaking Bad", "Stranger Things", "Suits"]
stars = [4.8, 3.2, 2.5, 3.0, 4.7, 3.8, 4.6, 3.1, 4.3]


plt.bar(show_categtories, stars, color="blue")
plt.xlabel("TV Shows / Movies")
plt.ylabel("Ratings")
plt.title("My ratings of Popular TV Shows and Movies")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

## Horizontal

```python
show_categtories = ["LOTR", "Harry Potter", "Star Wars", "GOT", "LOST", "Friends", "Breaking Bad", "Stranger Things", "Suits"]
stars = [4.8, 3.2, 2.5, 3.0, 4.7, 3.8, 4.6, 3.1, 4.3]


plt.barh(show_categtories, stars, color="blue")
plt.xlabel("TV Shows / Movies")
plt.ylabel("Ratings")
plt.title("My ratings of Popular TV Shows and Movies")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
