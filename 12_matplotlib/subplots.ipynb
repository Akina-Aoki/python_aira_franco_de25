# Subplots (Multiple Plots on One Figure)

# 1 Row, 2 Columns

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 points from 0 to 2 (for smooth curves)
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)    # Sine values for each angle in x
y2 = np.cos(x)    # Cosine values for each angle in x

# Create a figure and 2 subplots arranged in 1 row, 2 columns
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # figsize in inches (width, height)

# ----- Sine Plot -----
axs[0].plot(x, y1, color='navy', linewidth=2)   # Plot sine in first subplot
axs[0].set_title('Sine Wave')                   # Give this subplot a title
axs[0].set_xlabel('Angle [rad]')                # Label for x-axis
axs[0].set_ylabel('Amplitude')                  # Label for y-axis

# ----- Cosine Plot -----
axs[1].plot(x, y2, color='orange', linewidth=2)  # Plot cosine in second subplot
axs[1].set_title('Cosine Wave')                  # Title for subplot 2
axs[1].set_xlabel('Angle [rad]')                 # X-axis label

# Big main title above both subplots
plt.suptitle('Side-by-side Trig Fun', fontsize=14)

# tight_layout prevents overlapping labels/titles
plt.tight_layout()   # Adjust spacing to fit everything nicely

plt.show()           # Actually render the plots in the notebook or script
```

### 2 Rows, 1 Column
Use `plt.subplots()` for flexible grid layouts. Use `plt.subplot()` for quick, simple 1-off plots.


```python
# Generate random data for demonstration
data = np.random.randn(100)
fig, axs = plt.subplots(2, 1, figsize=(6, 8))
axs[0].hist(data, bins=20, color='g', alpha=0.7)
axs[0].set_title('Histogram')
axs[0].set_ylabel('Frequency')
axs[1].scatter(np.arange(100), data, color='r', marker='o')
axs[1].set_title('Scatter Plot')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Random Value')
plt.tight_layout()
plt.show()
```

## Customizing Style and Aesthetics

```python
plt.style.use('ggplot')  # Try 'seaborn', 'bmh', 'grayscale', etc.

x = np.arange(0, 10, 0.2)
plt.plot(x, np.sin(x), marker='o', label='sin(x)')
plt.plot(x, np.cos(x), marker='x', label='cos(x)')
plt.title('Styled Trig Functions')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

plt.style.use('default')  # Reset to Matplotlib's default style
```

## Adding Annotations
*Use `annotate()` to draw attention to key data points or features!*

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Annotated Sine Curve')
plt.xlabel('X')
plt.ylabel('sin(X)')
# Annotate the peak
plt.annotate(
    'First Peak',
    xy=(np.pi/2, 1),            # Location (data coords)
    xytext=(2, 1.3),            # Text location
    arrowprops=dict(facecolor='purple', arrowstyle='->'),
    fontsize=12,
    color='purple'
)
plt.show()
```

## Saving Figures
*Save in multiple formats for use in slides, papers, or sharing online.*

```python
plt.plot([1, 2, 3], [2, 4, 1], marker='s')
plt.title("Customized Save Example")
plt.grid(True)
plt.savefig('my_plot.png', dpi=150, bbox_inches='tight')  # Higher DPI for publication
plt.savefig('my_plot.pdf')      # Vector quality for print
plt.close()                     # Always close when done!
```

## Clearing and Closing
*Always manage memory when generating lots of plots or running in a loop!*

```python
# Suppose you make a quick plot, then want to reuse the figure slot
plt.plot([1,2,3], [4,5,6])
plt.clf()   # Clear current figure but dont close window

# To fully close and free memory (important if making many plots)
plt.close()

# For interactive loops/work, use plt.pause(0.01)
```
