# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sizes = [15, 30, 45, 10]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
explode = (0.1, 0, 0, 0)  # "Explode" the first slice to highlight Apples

plt.pie(
    sizes,
    explode=explode,         # slightly pop out the first slice
    labels=labels,           # slice labels
    autopct='%1.1f%%',       # display percentages
    shadow=True,             # drop shadow for 3D effect
    startangle=90            # rotate start so 'Apples' is at the top
)
plt.title('Fruit Popularity Pie Chart')
plt.axis('equal')  # Ensures pie is a circle, not an oval
plt.show()
```
