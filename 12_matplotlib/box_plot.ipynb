# Box Plot
*Boxplots are awesome for visualizing spread, outliers, and medians across categories!*

`np.random.normal(mean, standard_deviation, size)`<br>
 <br>
**means:**<br>
<br>
`mean` this is the mean (center of the curve). Most values will be around 6.

`standard_deviation` this is the standard deviation, which controls the spread of the data. Values cluster tightly around mean. values spread out more.

`size` this is the number of random samples to generate.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = np.random.normal(5, 1, 100)
data2 = np.random.normal(7, 2, 100)
data3 = np.random.normal(6, 1.5, 100)
data = [data1, data2, data3]

plt.boxplot(data,
            vert=True,         # vertical boxplots
            patch_artist=True, # color fills
            labels=['Exp1', 'Exp2', 'Exp3'])
plt.title('Boxplot of Three Experiments')
plt.ylabel('Measured Value')
plt.grid(True, axis='y')
plt.show()
```
