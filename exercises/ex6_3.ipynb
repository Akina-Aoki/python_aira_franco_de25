## Refactor the plot from matplotlib into a class

```python
# This should be the way to plot

# subplot() Just 1 figure in a plot
fig, ax = plt.subplots(1)
ax.plot(x, y)
ax.set(title = "Cool Stock", xlabel = "day", ylabel = "SEK")

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 
ax.annotate("Big Dip Here", xytext=(9, 10), xy=(7.5, 14),
            arrowprops=dict(arrowstyle="->"))

fig, ax = plt.subplots(1)

ax.plot(x, y)

annotator = PlotAnnotator(fig, ax)
annotator.arrow()
```

```python
class PlotAnnotator:
    def __init__(self, fig, ax):
        self.fig = fig
        self.ax = ax

    def arrow(self, label:str, start_xy:tuple, end_xy:tuple):
        self.ax.annotateax.annotate(
            label,
            start_xy,
            end_xy,
            xytext=(9, 10),
            xy=(7.5, 14),
            arrowprops=dict(arrowstyle="->")
            )
        
    def remove_clutter(self):
        self.ax.spines['top'].set_visible(False)
        self.ax.spines['right'].set_visible(False)

fig, ax = plt.subplots(1)

ax.plot(x, y)

annotator = PlotAnnotator(fig, ax)
annotator.arrow(label = "Big Dip Here", start_xy = (6, 10), end_xy = (7.5, 14.2))
annotator.remove_clutter()
```
